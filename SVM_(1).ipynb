```python
from sklearn.svm  import SVC 
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
ds = scipy.io.loadmat("ex6data1")
training_X = ds["X"]
training_r = ds["y"].reshape(len(ds["y"]))
svmclassifier = SVC(kernel='linear')
start = time.time()
svmclassifier.fit(training_X, training_r)
end = time.time()
print("Training time: ", end - start, " seconds")
training_y = svmclassifier.predict(training_X)
print("Trainig accuracy: ", np.mean(training_y == training_r) * 100, "%")
x1 = np.arange(0, 5, 0.01)
x2 = np.arange(0, 5, 0.01)
y = np.zeros((len(x1), len(x2)))

X1, X2 = np.meshgrid(x1, x2)

for ix1 in range(len(x1)):
    for ix2 in range(len(x2)):
        y[ix1, ix2] = svmclassifier.predict(np.array([[x1[ix1], x2[ix2]]]))
        
fig, ax = plt.subplots()
ax.plot(training_X[training_r == 1, 0], training_X[training_r == 1, 1], "k+")
ax.plot(training_X[training_r == 0, 0], training_X[training_r == 0, 1], "ro")

ax.contourf(X1.T, X2.T, y, cmap = plt.cm.winter)
plt.show()
```

```python
ds = scipy.io.loadmat("ex6data3")
training_X = ds["X"]
training_r = ds["y"].reshape(len(ds["y"]))
val_X = ds["Xval"]
val_r = ds["yval"].reshape(len(ds["yval"]))
svmclassifier = SVC(kernel='sigmoid', gamma = 90)
start = time.time()
svmclassifier.fit(training_X, training_r)
end = time.time()
print("Training time: ", end - start, " seconds")
training_y = svmclassifier.predict(training_X)
print("Trainig accuracy: ", np.mean(training_y == training_r) * 100, "%")
val_y = svmclassifier.predict(val_X)
print("Test accurasy: ",np.mean(val_y==val_r) * 100, "%")
x1 = np.arange(-0.7, 0.4, 0.01)
x2 = np.arange(-0.9, 0.8, 0.01)
y = np.zeros((len(x1), len(x2)))

X1, X2 = np.meshgrid(x1, x2)

for ix1 in range(len(x1)):
    for ix2 in range(len(x2)):
        y[ix1, ix2] = svmclassifier.predict(np.array([[x1[ix1], x2[ix2]]]))
        
fig, ax = plt.subplots()
ax.plot(training_X[training_r == 1, 0], training_X[training_r == 1, 1], "k+")
ax.plot(training_X[training_r == 0, 0], training_X[training_r == 0, 1], "ro")
ax.plot(val_X[val_r == 1, 0], val_X[val_r == 1, 1], "kx")
ax.plot(val_X[val_r == 0, 0], val_X[val_r == 0, 1], "rs")

ax.contourf(X1.T, X2.T, y, cmap = plt.cm.winter)
plt.show()
```
